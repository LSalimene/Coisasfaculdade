```
#Importar os pacotes que iremos utilizar
import pandas as pd
import matplotlib.pyplot as plt

#leitura do arquivo excel
data = pd.read_excel(open('../content/drive/My Drive/pelotas_temp.xlsx','rb'))
#aqui iremos definir nossas informações para facilitar o grafico
dia = pd.DataFrame(data,columns=['Dia'])
temp = pd.DataFrame(data,columns=['Temperatura media'])

#escolher o estilo que iremos usar
plt.style.use('ggplot')
#plotar o grafico e escolher a cor e o estilo de plot
plt.plot(dia,temp,'b-',dia,temp,'wo')
#nomear os eixos e dar o titulo
plt.xlabel('Dia')
plt.ylabel('Temperatura Media')
plt.title('Temperatura media primeira semana de janeiro')
#mostrar o grafico
plt.show()
```
